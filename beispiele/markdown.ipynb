# Tabellen

Eine Tabelle kann konstruiert werden mit dem | (Senkrechter Strich) für Spalten und - (Minuszeichen) unter den Überschriften. Der Text in den Spalten kann ausgerichtet werden mit :-: für zentriert, -: für rechtsbündig und :- für linksbündig (standard).


|Überschrift|Überschrift|Überschrift |
|:----------|:---------:|-----------:|
|linksbündig|zentriert  |rechtsbündig|
|A          |B          |C           |
|X          |Y          |Z           |
